```python
# Estimation/Data fit 

fecha = '2020-08-01'
#region = 'Nuevo_Leon'
region = 'Mexico'
#caso = "reported"
caso = "ocurred"

if caso == "reported":
    base_modelo = reportadas[(reportadas['date']>='2020-05-01') & (reportadas['date']<=fecha)].reset_index()
elif caso == "ocurred":
    base_modelo = ocurridas[(ocurridas['date']>='2020-05-01') & (ocurridas['date']<=fecha)].reset_index()
    

if region == 'Nuevo_Leon':
    P0 = 5120000
elif region == "Mexico":
    P0 = 126000000
    
I0 = base_modelo.loc[0, 'Cases_'+ region]
D0 = base_modelo.loc[0, 'Deaths_'+ region]


grid_ri = np.linspace(0.5, .2, num=10)
grid_rf = np.linspace(1, 1.2, num=10)
grid_lp = np.linspace(0.07, 0.3, num=10)
grid_sc = np.linspace(15, 10, num=10)


results = pd.DataFrame({'grid_sc':[],
                        'grid_ri': [],
                        'grid_rf': [],
                        'grid_lp': [],
                        'dist':    []})

indice = 0


for iri in tqdm(range(0,len(grid_ri))):
    for irf in range(0,len(grid_rf)):
        for ilp in range(0,len(grid_lp)):
            for sc in range(0,len(grid_sc)):
                
                
                Ri= grid_ri[iri]      # initial reproduction rate
                Rf= grid_rf[irf]      # final reproduction rate as if 0 infected: S/P=1
                λ = grid_lp[ilp]      # behavioral param
                sc = grid_sc[sc]     # scale parameter for initia infected


                %run -i "D:\francisco_madrigal\Desktop\ITAM\Semestre 3\Inv Aplicada I\modelos_SIR\BSIR_modelo.py"


                distc = (sum(np.power(
                    (
                        (df_cal[(df_cal['date']>='2020-05-01') & (df_cal['date']<=fecha)]['Deaths']) - 
                        (base_modelo[(base_modelo['date']>='2020-05-01') & (base_modelo['date']<=fecha)]['Deaths_'+region])
                    )
                    ,2)) / len(base_modelo))

                results.loc[indice,'grid_sc'] = sc
                results.loc[indice,'grid_ri'] = Ri
                results.loc[indice,'grid_rf'] = Rf
                results.loc[indice,'grid_lp'] = λ
                results.loc[indice,'dist'] = distc

                indice += 1
                        

min_val = results['dist'].idxmin()

print(f'{caso} - {region} - {fecha}:\n',
      'Suma de los errores promedio: {}\n'.format(results.loc[min_val,'dist']),
      'I0: {}, con un sc de {}\n'.format(results.loc[min_val,'grid_sc'] * I0, results.loc[min_val,'grid_sc']),
      'R0: {}\n'.format(results.loc[min_val,'grid_ri']),
      'Rf: {}\n'.format(results.loc[min_val,'grid_rf']),
      'λ: {}\n'.format(results.loc[min_val,'grid_lp']))

sc = results.loc[min_val,'grid_sc']
Ri= results.loc[min_val,'grid_ri']      # initial reproduction rate
Rf= results.loc[min_val,'grid_rf']      # final reproduction rate as if 0 infected: S/P=1
λ = results.loc[min_val,'grid_lp']      # behavioral param

%run -i "D:\francisco_madrigal\Desktop\ITAM\Semestre 3\Inv Aplicada I\modelos_SIR\BSIR_modelo.py"

tbound = 250
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_cal[df_cal.time<tbound]['date'],
    y=df_cal[df_cal.time<tbound]['Deaths'],
    name='Deaths ' + region + ' (Model)',
    mode='lines+markers',line_width=2,marker_size=6,opacity=1))

fig.add_trace(go.Scatter(
    x=base_modelo['date'],
    y=base_modelo['Deaths_' + region],
    name='Deaths ' + region + ' (Data as '+ caso + ')',
    mode='lines+markers',line_width=2,marker_size=6,opacity=1))

fig.update_layout(
    title="Model/Data " + region + " - deaths",
    title_x=0.5,showlegend=True,yaxis = {'type': 'linear'},
    legend=dict(
            x=0.10,y=0.95,
            traceorder="normal",font=dict(family="sans-serif",size=14),bgcolor='rgba(0,0,0,0)'),)

fig.show()

# Para estimar las máximas muertes diarias

df_cal['muertes_diarias'] = df_cal['Deaths'].diff()

print('Muertes despúes de 120 días (31 de diciembre de 2020): {}'.format(df_cal.loc[120,'Deaths']))
print('Mayor número de muertes: {}'.format(df_cal['muertes_diarias'].max()))
print('Días para alcanzar el mayor número de muertes: {}'.format(df_cal['muertes_diarias'].idxmax()))


res_agregados.loc[modelo,'region'] = region
res_agregados.loc[modelo,'caso'] = caso
res_agregados.loc[modelo,'fecha'] = fecha
res_agregados.loc[modelo,'I0'] = results.loc[min_val,'grid_sc'] * I0
res_agregados.loc[modelo,'Ri'] = results.loc[min_val,'grid_ri']
res_agregados.loc[modelo,'Rf'] = results.loc[min_val,'grid_rf']
res_agregados.loc[modelo,'lambda'] = results.loc[min_val,'grid_lp']
res_agregados.loc[modelo,'dist'] = results.loc[min_val,'dist']
res_agregados.loc[modelo,'muertes_120'] = df_cal.loc[120,'Deaths']
res_agregados.loc[modelo,'max_muertxdia'] = df_cal['muertes_diarias'].max()
res_agregados.loc[modelo,'dias_hasta_max_muertes'] = df_cal['muertes_diarias'].idxmax()
modelo += 1
```
